```python
import pandas as pd
```

```python
df=pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head()
```

```python
df['Job Title']
```

```python
df.info()
```

```python
#Removing -1 from the salary estimates
```

```python
df=df[df['Salary Estimate']!='-1']
```

```python
df.head()
```

```python
len(df)
```

```python
#Removing Glassdoor Est. from Salary Estimate 
```

```python
salary=df['Salary Estimate'].apply(lambda x:x.split('(')[0])
```

```python
salary
```

```python
kd_remove = salary.apply(lambda x:x.replace('K','').replace('$',''))
```

```python
kd_remove
```

```python
#adding new column to dataframe for hourly pay
```

```python
df['hourly']=df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df.head()
```

```python
min_hr=kd_remove.apply(lambda x:x.lower().replace('per hour','').replace('employer provided', '')) #removing perhour and employer provided from salary
```

```python
min_hr
```

```python
#splitting salary by max and min
```

```python
df['min_salary']=min_hr.apply(lambda x:x.replace(' salary:','')).apply(lambda x:int(x.split('-')[0]))
```

```python
df.head()
```

```python
df['max_salary']=min_hr.apply(lambda x:int(x.split('-')[1]))
```

```python
df.head() #min and max salaries added
```

```python
df.info()
```

```python
df['avg_salary']=(df['max_salary']+df['min_salary'])/2
```

```python
df.head()
```

```python
#Cleaning company names
```

```python
df['company_txt']=df.apply(lambda x:x['Company Name'] if x['Rating']<0 else x['Company Name'][:-4], axis=1)
```

```python
df.head()
```

```python
#Creating Job location
```

```python
df['Job state']=df['Location'].apply(lambda x:x.split(',')[1])
```

```python
df.drop(columns='Job Location', inplace=True)
```

```python
#Verifiying if jobs are at headquarters of the company
```

```python
df['same state']=df.apply(lambda x:1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
df.head()
```

```python
#calculating age, dataset from 2020
```

```python
df['age']=df.Founded.apply(lambda x: x if x<1 else 2020-x)
```

```python
#sparsing job description
#python
#rstudio
#spark
#aws
#excel
```

```python
df['Python']=df['Job Description'].apply(lambda x: 1 if 'python' in  x.lower() else 0)
```

```python
df['R Studio']=df['Job Description'].apply(lambda x: 1 if 'r studio' in  x.lower() or 'r-studio' in x.lower() else 0)
```

```python
df['Python'].value_counts()
```

```python
df['R Studio'].value_counts()
```

```python
df['Spark']=df['Job Description'].apply(lambda x: 1 if 'spark' in  x.lower() else 0)
```

```python
df['Spark'].value_counts()
```

```python
df['AWS']=df['Job Description'].apply(lambda x: 1 if 'aws' in  x.lower() else 0)
```

```python
df['AWS'].value_counts()
```

```python
df['excel']=df['Job Description'].apply(lambda x: 1 if 'excel' in  x.lower() else 0)
```

```python
df['excel'].value_counts()
```

```python
df.columns
```

```python
df.drop(columns='Unnamed: 0',inplace = True)
```

```python
df.head()
```

```python
df.to_csv('Cleaned dataset', index = False)
```
